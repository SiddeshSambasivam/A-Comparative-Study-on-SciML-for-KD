```python
import pandas as pd
import sympy
```

```python
equations = [
    "x_0**3+x_0**2+x_0",
    "x_0**4+x_0**3+x_0**2+x_0",
    "x_0**5+x_0**4+x_0**3+x_0**2+x_0",
    "x_0**6+x_0**5+x_0**4+x_0**3+x_0**2+x_0",
    "sin(x_0**2)*cost(x_0)-1",
    "sin(x_0)+sin(x_0+x_0**2)",
    "log(x_0+1)+log(x_0**2+1)",
    "sqrt(x_0)",
    "sin(x_0)+sin(x_1**2)",
    "2*sin(x_0)*cos(x_1)",
    "x_0**x_1",
    "x_0**4-x_0**3+0.5*x_1**2-x_1"
]

sampling_dist = [
    (-1, 1, 20, 1), # min, max, N, num_var
    (-1, 1, 20, 1),
    (-1, 1, 20, 1),
    (-1, 1, 20, 1),
    (-1, 1, 20, 1),
    (-1, 1, 20, 1),
    (0, 2, 20, 1),
    (0, 4, 20, 1),
    (0, 1, 20, 2),
    (0, 1, 20, 2),
    (0, 1, 20, 2),
    (0, 1, 20, 2),
]
```

```python
data = {
    "eq":[],
    "support":[],
    "num_points":[]
}

for eq, dist in zip(equations, sampling_dist):
    
    data["eq"].append(eq)
    a,b,n,v = dist
    
    var_dict = {}
    for i in range(v):
        var_dict[f'x_{i}'] = {"max":b, "min":a}
    
    data["support"] = str(var_dict)
    data["num_points"] = n

df = pd.DataFrame(data=data)
df
```
